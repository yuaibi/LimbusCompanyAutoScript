```python
import win32gui
import win32con
import win32ui
import numpy as np

# 初始化窗口捕获，返回窗口设备上下文和内存设备上下文
def initialize_window_capture(window_name):
    # 获取窗口句柄
    hwnd = win32gui.FindWindow(None, window_name)
    if hwnd == 0:
        print(f"Window '{window_name}' not found!")
        return None, None, None
    
    # 获取窗口位置
    left, top, right, bottom = win32gui.GetClientRect(hwnd)

    # 获取窗口的宽高
    width = right - left
    height = bottom - top
    
    # 获取窗口设备上下文
    hwindc = win32gui.GetWindowDC(hwnd)
    srcdc = win32ui.CreateDCFromHandle(hwindc)
    
    # 创建内存设备上下文
    memdc = srcdc.CreateCompatibleDC()
    
    # 返回初始化结果
    return hwnd, srcdc, memdc, width, height

# 截取窗口图像并保存为文件
def capture_window_image(hwnd, srcdc, memdc, width, height):
    # 恢复窗口（如果窗口最小化了）
    if win32gui.IsIconic(hwnd):  # 如果窗口是最小化的
        win32gui.ShowWindow(hwnd, win32con.SW_RESTORE)  # 恢复窗口
    
    # 创建位图对象
    bitmap = win32ui.CreateBitmap()
    bitmap.CreateCompatibleBitmap(srcdc, width, height)
    
    # 将窗口内容复制到内存位图
    memdc.SelectObject(bitmap)
    memdc.BitBlt((0, 0), (width, height), srcdc, (0, 0), win32con.SRCCOPY)
    
    # 获取位图内容并转换为 PIL 图像
    signed_data = bitmap.GetBitmapBits(True)
    img = np.frombuffer(signed_data, dtype=np.uint8)
    img.shape = (height, width, 4)  # 可能需要调整颜色格式
    
    # 去除透明度通道，只保留 BGR
    img = img[..., :3]  # 去掉 alpha 通道（透明度）
    
    # 将 BGR 转换为 RGB
    img_rgb = img[..., ::-1]  # 反转通道顺序（BGR -> RGB）

    # 清理
    win32gui.DeleteObject(bitmap.GetHandle())
    memdc.DeleteDC()
    srcdc.DeleteDC()
    win32gui.ReleaseDC(hwnd, win32gui.GetWindowDC(hwnd))

    return img_rgb
```

```python
import time
import psutil
import threading
from datetime import datetime

# 假设 capture_window 函数已经定义好

# 测试截图函数的时间消耗、长时间稳定性和资源消耗
def test_capture_window(window_name, num_tests):

    # 记录开始时间
    start_time = time.time()
    capture_times = []  # 存储每次截图的时间消耗

    # 获取进程ID和进程对象（用于监控资源）
    process = psutil.Process()

    print("测试开始时间:", datetime.now())

    #初始化capture_window
    

    # 循环多次执行截图
    for i in range(num_tests):
        print(f"开始第 {i+1} 次截图...")
        # 测量每次截图的时间消耗
        capture_start_time = time.time()

        # 调用 capture_window 函数进行截图
        hwnd, srcdc, memdc, width, height = initialize_window_capture(window_name)
        if hwnd == 0:
            print(f"Window '{window_name}' not found!")
            print(f'初始化失败')
            return 
        img = capture_window_image(hwnd, srcdc, memdc, width, height)
        plt.imshow(img)
        plt.show()

        # 记录结束时间并计算消耗的时间
        capture_end_time = time.time()
        capture_duration = capture_end_time - capture_start_time
        capture_times.append(capture_duration)
        
        # 打印每次截图的时间消耗
        print(f"第 {i+1} 次截图耗时: {capture_duration:.4f} 秒")

    # 记录结束时间
    end_time = time.time()

    # 计算总的测试时间
    total_time = end_time - start_time
    print(f"\n总测试时间: {total_time:.4f} 秒")
    
    # 输出平均截图时间
    avg_capture_time = sum(capture_times) / len(capture_times)
    print(f"平均截图时间: {avg_capture_time:.4f} 秒")

    # 监控 CPU 和内存消耗
    print("\n资源消耗分析:")
    cpu_percent = process.cpu_percent(interval=1)  # 获取 CPU 使用率
    memory_info = process.memory_info()  # 获取内存信息
    print(f"CPU 使用率: {cpu_percent}%")
    print(f"内存使用情况: {memory_info.rss / (1024 * 1024):.2f} MB")

    print("\n测试完成！")


# 使用多线程来模拟长时间运行中的稳定性
def run_test_in_thread(window_name, num_tests):
    test_thread = threading.Thread(target=test_capture_window, args=(window_name, num_tests))
    test_thread.start()
    test_thread.join()

    print("\n所有测试完成！")
```

```python
# 执行多个线程的测试
window_name = "MuMu安卓设备"
num_tests = 100
test_capture_window(window_name, num_tests)
```
